```python
import pandas as pd

df = pd.read_excel(r"C:\Users\Rutvi\OneDrive\Desktop\Streaming Churn Intelligence Behavioral Segmentation & Retention Analytics\archive (6)\netflix_large_user_data.xlsx")

df_fe = df.copy()
```

```python
if 'watch_time' in df_fe.columns and 'sessions' in df_fe.columns:
    df_fe['engagement_score'] = df_fe['watch_time'] / (df_fe['sessions'] + 1)
```

```python
if 'watch_time' in df_fe.columns and 'sessions' in df_fe.columns:
    df_fe['binge_score'] = df_fe['watch_time'] * df_fe['sessions']
```

## 📊 Activity Level Segmentation

We group users into:
- Low
- Medium
- High engagement

```python
if 'engagement_score' in df_fe.columns:
    df_fe['activity_level'] = pd.cut(
        df_fe['engagement_score'],
        bins=3,
        labels=['Low', 'Medium', 'High']
    )
```

## ⚠️ Identify High Risk Users

We flag users who are low engagement and churned.

```python
target_candidates = [col for col in df_fe.columns 
                     if 'churn' in col.lower() 
                     or 'cancel' in col.lower() 
                     or 'active' in col.lower()]

target_col = target_candidates[0] if len(target_candidates) > 0 else None

if target_col and 'engagement_score' in df_fe.columns:
    df_fe['high_risk_user'] = (
        (df_fe['engagement_score'] < df_fe['engagement_score'].median()) &
        (df_fe[target_col] == 1)
    )
```
